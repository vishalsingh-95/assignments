# Q1

# the five ploting functions we use are:-

# 1) Scatter Plot:
It is used to visualize the relationship between two numerical variables. It helps identify patterns, clusters, or correlations in the data.

# 2) Line Plot:
It is useful for visualizing the trend or evolution of a variable over time. It is commonly used to show time series data or continuous data.

# 3) Bar Plot:
It is effective in comparing categorical data by displaying the height of rectangular bars. It helps visualize the distribution, comparison, or composition of different categories.

# 4) Histogram:
It displays the distribution of a single numerical variable by dividing it into bins and counting the frequency of values falling into each bin. It helps understand the data's shape, central tendency, and spread.

# 5) Box Plot:
It provides a summary of the distribution of a numerical variable through quartiles and identifies outliers. It helps visualize the median, range, and skewness of the data, and detect potential anomalies.

# Q2

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(data=fmri,x=fmri.timepoint,y=fmri.signal,hue=fmri.event,style=fmri.region)
plt.show()
```

# Q3

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.title('Box Plot Age v/s Pclass')
plt.show()
```

```python
sns.boxplot(x=titanic.pclass,y=titanic.fare)
plt.title('Box Plot Fare v/s Pclass')
plt.show()
```

Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
```

```python
flights_data=flights.pivot("month","year","passengers")
```

```python
flights_data
```

# Matplotlib is a popular data visualization library in Python that provides a wide range of plotting functionalities. It is used to create high-quality plots, charts, and figures to visualize data in a visually appealing and informative manner.

# Five plots that can be plotted using the Pyplot module are:-
1) Line Plot: A line plot is created by connecting data points with straight line segments. It is useful for showing trends or changes over time.
2) Scatter Plot: A scatter plot represents individual data points as markers on a two-dimensional plane. 
3) Bar Plot: A bar plot displays rectangular bars whose lengths represent the values of different categories
4) Histogram: A histogram is a graphical representation of the distribution of numerical data. It consists of bars where the height represents the frequency or count of data falling into specific bins or intervals.
5) Pie Chart: A pie chart is a circular statistical graphic divided into sectors, where the area of each sector represents the proportionate value of a category. It is commonly used to show the composition or distribution of a whole into its parts.

# Scatter Plot: A scatter plot represents individual data points as markers on a two-dimensional plane.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.title('scatter_plot')
plt.scatter(x,y)
plt.show()
```

# The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple subplots and specify the layout using rows, columns, and the index of each subplot.



```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3= np.array([10, 20, 30, 40, 50, 60])
x4= np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("line1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("line2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("line3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("line4")
plt.show()
```

```python
# What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
```

# A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays key statistical measures such as the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum.

# The uses of boxplot() are:-
1) Summary of Distribution: A box plot provides a concise visual summary of the distribution of a dataset, giving insights into its central tendency, spread, skewness, and presence of outliers. It allows for quick comparisons of multiple distributions.
2) Identification of Outliers: Outliers are data points that significantly deviate from the rest of the dataset.
3) Comparison between Groups: Box plots are particularly useful for comparing distributions between different groups or categories. By plotting multiple box plots side by side, it becomes easy to compare their medians, ranges, and quartiles, revealing differences or similarities in the data.
4) Detection of Skewness: The shape of the box plot can indicate the presence of skewness in the dataset. Skewness refers to the asymmetry of the distribution.
5) Visualizing Data Distribution: The box plot can provide a visual representation of the distribution of the dataset, helping to assess the concentration of data points around the median and quartiles. It can reveal if the data is symmetric, skewed, or has multiple peaks.

```python
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('Boxplot')
plt.boxplot(data)
plt.show()
```
